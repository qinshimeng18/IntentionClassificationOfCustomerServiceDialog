### Word Embedding

```python
#try cut word and word embedding by own corpus
import jieba
from gensim.models.word2vec import Word2Vec  
```

```python
# split the word and build the corpus
corpus = []
for i in df2.response:
    corpus.append([word for word in jieba.cut(i)])
for i in df2.request:
    corpus.append([word for word in jieba.cut(i)])
```

```python
model=Word2Vec()  
model.build_vocab(corpus)  
model.train(corpus,total_examples=model.corpus_count,epochs=model.iter)  
```

```python
model.save('self_train_word2vec')  
# model.save_word2vec_format('text.model.bin', binary=True)
```
